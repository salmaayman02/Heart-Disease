```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
```

```python
data['HeartDisease'].value_counts(normalize= True)*100
```

```python
data['HeartDisease'].value_counts().sort_index().plot.bar(
    color = 'teal',
    edgecolor = 'black',
    figsize = (8,6)
)

plt.title('Patient distribution', fontsize=16, fontweight='bold')
plt.xlabel('HeartDisease', fontsize=14)
plt.ylabel('Number of Patient', fontsize=14)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(rotation=45)  
plt.tight_layout()      
plt.show()
```

```python
pd.crosstab(data.HeartDisease,data.Smoking)
```

```python
# Create the crosstab and plot it as a bar chart
pd.crosstab(data.HeartDisease,data.Smoking).plot.bar(
    figsize=(10, 6),
    color=['#41b6c4', '#08306b'],
    edgecolor='black'
)

# Add titles and labels
plt.title('Heart Disease and Smoking', fontsize=16)
plt.xlabel('Heart Disease ', fontsize=14)
plt.ylabel('Frequnecy', fontsize=14)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust legend for clarity
plt.legend(title='Smoking', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```
